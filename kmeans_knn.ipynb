```python
from sklearn.datasets import load_iris
data = load_iris()
X = data.data[:,2:4]
y = data.target
data.target_names
```

```python
import matplotlib.pyplot as plt

for i in range(len(set(y))):
  plt.plot(X[y==i, 0], X[y==i, 1], ".", label=data.target_names[i])
plt.xlabel("Petal lenght")
plt.ylabel("Petal width")
plt.legend()
plt.show()
```

```python
plt.scatter(X[:,0], X[:,1], marker=".")
plt.xlabel("Petal lenght")
plt.ylabel("Petal width")
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np

kmeans = KMeans(n_clusters=3, n_init="auto", random_state=2).fit(X)
print(kmeans.cluster_centers_)

plt.scatter(X[:,0], X[:,1], marker=".")
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], marker="x", color="r", s=100, linewidths=3)
plt.xlabel("Petal lenght")
plt.ylabel("Petal width")
plt.show()
```

```python
for i in range(len(set(y))):
  plt.plot(X[y==i, 0], X[y==i, 1], "s", label=data.target_names[i])
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], marker="x", color="r", s=100, linewidths=3)
ypred = kmeans.predict(X)
for i in range(len(set(ypred))):
  plt.plot(X[ypred==i, 0], X[ypred==i, 1], ".", label=data.target_names[i])
plt.xlabel("Petal lenght")
plt.ylabel("Petal width")
plt.legend()
plt.show()
```

KNeighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X, y)
for i in range(len(set(y))):
  plt.plot(X[y==i, 0], X[y==i, 1], "s", label=data.target_names[i])
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], marker="x", color="r", s=100, linewidths=3)
ypred_knn = knn.predict(X)
for i in range(len(set(ypred_knn))):
  plt.plot(X[ypred_knn==i, 0], X[ypred_knn==i, 1], ".", label=data.target_names[i])
plt.xlabel("Petal lenght")
plt.ylabel("Petal width")
plt.legend()
plt.show()
```

```python
for i in range(len(set(ypred))):
  plt.plot(X[ypred==i, 0], X[ypred==i, 1], "s", label=data.target_names[i])
for i in range(len(set(ypred_knn))):
  plt.plot(X[ypred_knn==i, 0], X[ypred_knn==i, 1], ".", label=data.target_names[i])
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], marker="x", color="r", s=100, linewidths=3)
plt.xlabel("Petal lenght")
plt.ylabel("Petal width")
plt.legend()
plt.show()
```
